```python
sales = pd.read_csv(r'C:\Users\ANGGARA MAHESWARA\Documents\semester 5\Data Mining\data_penjualan.csv', sep=';')
```

```python
# Bersihkan kolom data dan ubah tipe data yang diperlukan
def clean_column(col):
    return col.astype(str).str.replace('[^0-9]', '', regex=True).astype(float)

sales['Jumlah Order'] = clean_column(sales['Jumlah Order'])
sales['Total'] = clean_column(sales['Total'])
sales['Tanggal'] = pd.to_datetime(sales['Tanggal'], dayfirst=True)
```

```python
# Menambahkan kolom 'No' untuk identifikasi unik
sales['No'] = sales.index
```

```python
# Buat pivot table per tanggal, jenis produk, dan harga
daily_sales = sales.pivot_table(
    index=['No', 'Tanggal', 'Jenis Produk', 'Harga'],
    values=['Jumlah Order', 'Total'],
    aggfunc='sum'
).reset_index()
```

```python
# Fungsi pencarian produk tertentu dan plot data
def filter_and_plot(product_name):
    filtered_data = daily_sales[daily_sales['Jenis Produk'].str.contains(product_name, case=False, regex=False)]
    if not filtered_data.empty:
        filtered_data.plot(x='Tanggal', y='Jumlah Order', kind='line', title=f"Tren Penjualan {product_name}")
        plt.show()
```

```python
# Analisis beberapa produk
for product in ["dupleks 270", "dupleks 310", "dupleks 350", "ivory 230", "ivory 250", "ivory 270", "ivory 300"]:
    filter_and_plot(product)
```
